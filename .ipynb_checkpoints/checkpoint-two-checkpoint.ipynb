```python
# import dependencies 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
# Read the dataset and Create dataframe
fuel_consumption_df= pd.read_csv("MY2022 Fuel Consumption Ratings.csv")
fuel_consumption_df.head()
```

```python
#Drop the Fuel Consumption(Comb (mpg)) and Model_year field, since it have only combined mileage values while L/km have all city, highway and combined values.
fuel_consumption_df = fuel_consumption_df.drop(['Fuel Consumption(Comb (mpg))','Model Year'], axis=1)
```

```python
# Check the count
fuel_consumption_df.count()
```

```python
# Drop the duplicate records from the dataset
fuel_consumption_df.drop_duplicates().count()
```

```python
fuel_consumption_df.columns
```

```python
# Find the null values in the dataset
fuel_consumption_df.isnull().sum()
```

```python
#Combine Make and Model to one field
fuel_consumption_df['Make_Model'] = fuel_consumption_df['Make'] +'_'+ fuel_consumption_df['Model']
```

```python
fuel_consumption_df.head()
```

```python
#Write this dataframe to csv file
fuel_consumption_df.to_csv('fuel_consumption_df.csv',index=False)
```

```python
#Find out number of models each make have
No_of_models_by_brand = fuel_consumption_df[['Make','Model']].drop_duplicates().Make.value_counts()\
.reset_index(name='Total models by Make').rename(columns={"index" : "Make"})
No_of_models_by_brand
```

```python
#Findout the number of models by each make and vehicle class
fuel_consumption_df.groupby(['Make','Vehicle Class'], as_index = False)['Model'].count().rename(columns={"Model" : "Number of Models"}).sort_values(by=['Make'])
#.sort_values(by=['Number of Models'],ascending = False)
```

```python
# Find out average combined fuel consumption in litres for all car makes
Combined_avg_fuel_consumption_by_make_model =fuel_consumption_df.groupby(['Make_Model'], as_index = False)['Fuel Consumption(Comb (L/100 km))'].mean().sort_values(by=['Fuel Consumption(Comb (L/100 km))'])
```

```python
Combined_avg_fuel_consumption_by_make_model.head()
```

```python
# Find out average CITY fuel consumption in litres for all car makes
city_avg_fuel_consumption_by_make_model =fuel_consumption_df.groupby(['Make_Model'], as_index = False)['Fuel Consumption (City (L/100 km)'].mean().sort_values(by=['Fuel Consumption (City (L/100 km)'])
```

```python
city_avg_fuel_consumption_by_make_model.head()
```

```python
# Find out average HWY fuel consumption in litres for all car makes
hwy_avg_fuel_consumption_by_make_model =fuel_consumption_df.groupby(['Make_Model'], as_index = False)['Fuel Consumption(Hwy (L/100 km))'].mean().sort_values(by=['Fuel Consumption(Hwy (L/100 km))'])
```

```python
hwy_avg_fuel_consumption_by_make_model.head()
```

```python
# Find out average combined fuel consumption in litres by make and vehicle class 
fuel_consumption_by_make_class =fuel_consumption_df.groupby(['Make_Model','Vehicle Class','Fuel Type'], as_index = False)['Fuel Consumption (City (L/100 km)','Fuel Consumption(Hwy (L/100 km))','Fuel Consumption(Comb (L/100 km))'].sum().sort_values(by=['Fuel Consumption (City (L/100 km)','Fuel Consumption(Hwy (L/100 km))','Fuel Consumption(Comb (L/100 km))'])
```

```python
fuel_consumption_by_make_class.head()
```

```python
# Find out total combined fuel consumption by make,model, class and Fuel type
fuel_consumption_by_make_Model_class_fuel =fuel_consumption_df.groupby(['Make','Model','Vehicle Class','Fuel Type'], as_index = False)['Fuel Consumption (City (L/100 km)','Fuel Consumption(Hwy (L/100 km))','Fuel Consumption(Comb (L/100 km))'].sum().sort_values(by=['Fuel Consumption (City (L/100 km)','Fuel Consumption(Hwy (L/100 km))','Fuel Consumption(Comb (L/100 km))'])
```

```python
# Create a list with unique values of vehicle class
vc_list =fuel_consumption_df['Vehicle Class'].unique().tolist()
# Create  an empty dataframe
best_mileage_by_make_model_vehicle_class = pd.DataFrame()
#Run a loop through vehicle class to find out best mileage car model by each vehicle class
for vclass in vc_list:
    temp_ps = fuel_consumption_df[fuel_consumption_df['Vehicle Class'] == vclass].sort_values(by=['Fuel Consumption(Comb (L/100 km))']).iloc[0]
    best_mileage_by_make_model_vehicle_class = best_mileage_by_make_model_vehicle_class.append(temp_ps,ignore_index=True)
    
best_mileage_by_make_model_vehicle_class    
```

```python
# Create  an empty dataframe
bad_mileage_by_make_model_class = pd.DataFrame()
#Run a loop through vehicle class to find out bad mileage car model in each vehicle class
for vclass in vc_list:
    temp_ps = fuel_consumption_df[fuel_consumption_df['Vehicle Class'] == vclass].sort_values(by=['Fuel Consumption(Comb (L/100 km))'],ascending = False).iloc[0]
    bad_mileage_by_make_model_class = bad_mileage_by_make_model_class.append(temp_ps,ignore_index=True)
    
bad_mileage_by_make_model_class    
```

```python
# Create a list with unique values of Fuel Type
ft_list =fuel_consumption_df['Fuel Type'].unique().tolist()
# Create  an empty dataframe
best_mileage_cars_in_each_fuel_type = pd.DataFrame()
#Run a loop through Fuel type to find out best mileage car model in each Fuel type
for ft in ft_list:
    temp_ps = fuel_consumption_df[fuel_consumption_df['Fuel Type'] == ft].sort_values(by=['CO2 Rating'],ascending = False).iloc[0]
    best_mileage_cars_in_each_fuel_type = best_mileage_cars_in_each_fuel_type.append(temp_ps,ignore_index=True)

best_mileage_cars_in_each_fuel_type    
```

```python
# Create  an empty dataframe
best_cars_with_smog_rating_by_class = pd.DataFrame()
#Run a loop through Fuel Type to find out best mileage car model in each Fuel type
for ft in vc_list:
    temp_ps = fuel_consumption_df[fuel_consumption_df['Vehicle Class'] == ft].sort_values(by=['Smog Rating'],ascending = False).iloc[0]
    best_cars_with_smog_rating_by_class = best_cars_with_smog_rating_by_class.append(temp_ps,ignore_index=True)
best_cars_with_smog_rating_by_class    
```
